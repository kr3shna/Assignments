Qs:n:
Show the year difference between the first and last medal won by each country

```python

dfn = df.groupby('team')['year'].agg(['min', 'max'])
dfn['Year_Difference'] = dfn['max'] - dfn['min']
```

```python
dfn
```

Qs:t:
Find athletes who won medals in three consecutive Olympic games in the same event

```python

dft = df.groupby(['name', 'sport', 'year']).size()
dftt = dft[dft >= 3].reset_index()['name'].unique()
```
